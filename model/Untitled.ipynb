```python
df["state"] = "All"
df["min_income"] = 0
df["max_income"] = 500000
df["min_age"] = 18
df["max_age"] = 60
df["category"] = "All"
df["target_group"] = "General"
df["deadline"] = "2026-12-31"
```

```python
df.loc[df["scheme_name"].str.contains("Mahila|Women", case=False), 'target_group'] = 'Women'
df.loc[df['scheme_name'].str.contains("Mahila|Women", case=False), 'max_age'] = 50
df.loc[df["scheme_name"].str.contains("Student|Training|Education", case=False), "target_group"] = "Students"
df.loc[df["scheme_name"].str.contains("Student|Training|Education", case=False), "min_age"] = 17
df.loc[df["scheme_name"].str.contains("Student|Training|Education", case=False), "max_age"] = 25
df.loc[df["scheme_name"].str.contains("Student|Training|Education", case=False), "max_income"] = 300000
df.loc[df["scheme_name"].str.contains("SC", case=False), "category"] = "SC"
df.loc[df["level"] == "State", "state"] = "Maharashtra"
df.head()
```

```python
df
```

```python
from datetime import datetime

#test user
user = {
    "age": 22,
    "income": 250000,
    "state": "Maharashtra",
    "category": "All"
}
#score funtion
def calculate_score(user, scheme):
    score = 0
    reasons = []

    if scheme["min_income"] <= user["income"] <= scheme["max_income"]:
        score += 40
        reasons.append("Income matched")

    if scheme["state"] == "All" or scheme["state"] == user["state"]:
        score += 30
        reasons.append("State matched")

    if scheme["category"] == "All" or scheme["category"] == user["category"]:
        score += 20
        reasons.append("Category matched")

    if scheme["min_age"] <= user["age"] <= scheme["max_age"]:
        score += 10
        reasons.append("Age matched")

    return score, reasons

#deadline alert

def check_deadline_alert(deadline_str):
    today = datetime.today()
    deadline = datetime.strptime(deadline_str, "%Y-%m-%d")
    
    days_left = (deadline - today).days
    
    if days_left < 0:
        return "Deadline Passed"
    elif days_left <= 30:
        return f"Deadline approaching in {days_left} days"
    else:
        return f"{days_left} days remaining"

recommendations = []

for _, scheme in df.iterrows():
    score, reasons = calculate_score(user, scheme)

    recommendations.append({
        "scheme_name": scheme["scheme_name"],
        "score": score,
        "status": classify(score),
        "reasons": ", ".join(reasons),
        "benefits": scheme["benefits"],
        "documents": scheme["documents"],
        "deadline": scheme["deadline"],
        "alert": check_deadline_alert(scheme["deadline"])
    })

rec_df = pd.DataFrame(recommendations)
rec_df = rec_df.sort_values(by="score", ascending=False)

#classification function
def classify(score):
    if score >= 80:
        return "Highly Eligible"
    elif score >= 50:
        return "Potentially Eligible"
    else:
        return "Not Eligible"

rec_df["status"] = rec_df["score"].apply(classify)


rec_df.head()
```
